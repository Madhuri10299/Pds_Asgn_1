**BMI Calculation**

```python
# Calculate BMI from weight & height

data['bmi'] = (703 * (data['weight']))/(data['height'] ** 2)
data['bmi'] = data['bmi'].round(2)

# Calculate grip strength in pounds
data['grip_strength_lbs'] = data['grip strength'] * 2.20462  # Conversion factor from KGS to LBS
data['grip_strength_lbs'] = data['grip_strength_lbs'].round(2)
# Calculate frailty index
# Got this from open source
data['frailty_index'] = (data['height'] / data['weight']) * data['age'] + (data['grip_strength_lbs'] / data['weight']) * data['age']
data['frailty_index'] = data['frailty_index'].round(2)

# Display the DataFrame with frailty index
print(data)
```

```python
data.to_csv("/content/cleaned_women_frailty1.csv")
```

**DATA VISUALIZATION**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and it contains 'age', 'frailty_index', and a categorical 'frailty' column

# Create a boxen plot of frailty_index across different frailty categories
sns.catplot(x='frailty', y='frailty_index', kind='boxen', data=data, palette='viridis', height=6, aspect=1.5)

plt.suptitle('Frailty Index across Frailty Categories', fontsize=16, y=1.05)
plt.subplots_adjust(top=0.9)  # Adjust the top to make space for suptitle

# Adjust the file save path to work in this environment
plt.savefig('/content/Snapshots/Frailty_Index_vs_Frailty_Category_BoxenPlot.jpg')
plt.show()
```

**BMI Vs Frailty status**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Loop through each frailty status to create a KDE plot
for frailty_status in data['frailty'].unique():
    subset = data[data['frailty'] == frailty_status]
    sns.kdeplot(subset['bmi'], label=str(frailty_status), fill=True, alpha=0.4)

plt.title('KDE of BMI by Frailty Status')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.legend(title='Frailty Status')
plt.grid(True)

# Adjust the file save path to work in this environment
plt.savefig('/content/Snapshots/bmi_vs_frailty_status_kde.png')
plt.show()
```

**Average of Grip strength by Frailty Status**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
# Violin plot
sns.violinplot(x='frailty', y='bmi', data=data, palette='Set2', inner="quartile")

plt.title('BMI by Frailty Status with Distribution Shape')
plt.xlabel('Frailty Status')
plt.ylabel('BMI')
plt.grid(True)

# Adjust the file save path to work in this environment
plt.savefig('/content/Snapshots/bmi_vs_frailty_violinplot.png')
plt.show()
```
